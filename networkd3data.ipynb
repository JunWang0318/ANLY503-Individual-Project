```python
import pandas as pd
import numpy as np
nfa = pd.read_csv('NFA.csv')
nfa.head()
```

```python
sub = nfa.loc[(nfa.UN_region=='Asia')& (nfa.year==2014) & (nfa.record=='EFConsPerCap')]
sub.head()
```

```python
network = sub.drop(columns = ['ISO alpha-3 code','UN_region','UN_subregion','year','record',
                              'Percapita GDP (2010 USD)', 'population'])
network.dropna()
```

```python
def normalize(df,cols):
    result = df.copy()
    for feature_name in cols:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
cols = ['crop_land','grazing_land', 'forest_land', 'fishing_ground', 'built_up_land', 'carbon','total']
norm_net = normalize(network,cols)
norm_net.head()
```
