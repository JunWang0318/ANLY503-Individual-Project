```python
import pandas as pd
import numpy as np
import plotly.plotly as py
from plotly.graph_objs import *
#from plotly import __version__
#from plotly.offline import download_plotlyjs, init_notebook_mode, plot,iplot
#print (__version__) # requires version >= 1.9.0
import plotly.graph_objs as go

df = pd.read_csv('/Users/jun/Desktop/503 final/NFA.csv')
EFConsPerCap = df.loc[df['record'] == "EFConsPerCap"]
Mexico = EFConsPerCap.loc[df['country'] == 'Mexico']
China = EFConsPerCap.loc[df['country'] == 'China']
UK = EFConsPerCap.loc[df['country'] == 'United Kingdom']
USA = EFConsPerCap.loc[df['country'] == 'United States of America']
Australia = EFConsPerCap.loc[df['country'] == 'Australia']
```

```python
Mexico = go.Scatter(
    x=Mexico['year'],
    y=Mexico['total'],
    name='Mexico'
)


China = go.Scatter(
    x=China['year'],
    y=China['total'],
    name='China'
)


UK = go.Scatter(
    x=UK['year'],
    y=UK['total'],
    name='United Kingdom'
)

USA = go.Scatter(
    x=USA['year'],
    y=USA['total'],
    name='United States of America'
)

Australia = go.Scatter(
    x=Australia['year'],
    y=Australia['total'],
    name='Australia'
)

data = [Australia, China, Mexico,UK, USA]

layout = go.Layout(
    title='Ecological Footprint of Consumption (percapita) for Five Representative Countries',
    xaxis=dict(
        title='Year',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Ecological Consumption Percapita (gha/person)',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

fig = go.Figure(data=data, layout=layout)
plot_url = py.plot(fig, filename='fivecountries_linechart')
```
